```python
import pandas as pd
```

```python
user = pd.read_csv('../../clean_data/user_study.csv')
```

```python
def get_decade(year):
    return f'{(year // 10) * 10}s'
    
# 保留具有生日信息的用户
user_birthday_df = user.dropna(subset=['birthday']).copy()
user_birthday_df['birthday'] = pd.to_datetime(
    user_birthday_df['birthday'], 
    format='%Y-%m-%d', 
    errors='coerce'  
)
user_birthday_df = user_birthday_df[user_birthday_df['birthday'].notna()]
# 提取年份并计算年代
user_birthday_df['decade'] = user_birthday_df['birthday'].dt.year.apply(get_decade)

# 清洗不可信的年代数据
user_birthday_df['year'] = user_birthday_df['birthday'].dt.year
start_year = 1940
end_year = pd.Timestamp.now().year - 3
user_birthday_df['is_valid'] = (user_birthday_df['year'] >= start_year) & (user_birthday_df['year'] <= end_year)
user_valid_data = user_birthday_df[user_birthday_df['is_valid']]
del user_valid_data['year']
del user_valid_data['nickname']
del user_valid_data['userid']
del user_valid_data['listen_song_num']
del user_valid_data['fan_count']
del user_valid_data['follow_count']
del user_valid_data['event_count']
print('有生日信息的用户数:', len(user_valid_data))

user_valid_data
```

## 年龄分组

```python
decade_counts = user_valid_data.groupby('decade').size().reset_index(name='count')
decade_counts
```
